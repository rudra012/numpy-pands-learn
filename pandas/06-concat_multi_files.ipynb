```python
from glob import glob
stock_files = sorted(glob('data/stocks*.csv'))
stock_files
```

```python
# This will concat file as per rows
pd.concat((pd.read_csv(file) for file in stock_files)).head()
```

```python
pd.concat((pd.read_csv(file) for file in stock_files), ignore_index=True).head()
```

```python
# To concat file as per columns
drink_files = sorted(glob('data/drinks*.csv'))
pd.concat((pd.read_csv(file) for file in drink_files), axis='columns').head()
```
